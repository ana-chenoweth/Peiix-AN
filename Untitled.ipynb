````python
from ollama import Client
import pandas as pd

# Cliente Ollama (el modelo ya corre en localhost:11434)
client = Client()

# Prompt del sistema (muy importante)
SYSTEM_PROMPT = """
Eres un agente experto en pandas. Responde SOLO una expresi√≥n Python v√°lida y ejecutable con eval().

REGLAS OBLIGATORIAS:

1. SOLO usa estas columnas EXACTAS, NO inventes otras:
   ["Reader serial number", "ID Merchant", "Merchant name", "State",
    "Business line", "Transaction date", "ID transaction", "Card type",
    "Card brand", "TPV", "Comission", "Txn type", "Ubication",
    "Capture", "ID Distributor", "Payment to distrubutor", "MIF"]

2. NUNCA generes columnas como:
   "Line of business", "Merchant", "Commission", "Mastercard",
   o cualquier variante que no exista.

3. NUNCA uses df.query().

4. NUNCA uses asignaciones: no escribas df = ...

5. NUNCA uses eval(), exec(), import, lambda.

6. TODAS las operaciones deben ser con sintaxis pandas est√°ndar:
   df[(cond1) & (cond2) & (cond3)]

7. Para fechas, SIEMPRE asume que:
   df["Transaction date"] ya es datetime.

   Ejemplos v√°lidos:
   df["Transaction date"].dt.day_name()
   df["Transaction date"].dt.hour

8. Si necesitas filtrar por tipo de tarjeta:
   - Usa EXACTAMENTE: "CREDIT" o "DEBIT"

9. Si necesitas filtrar por marca:
   Usa EXACTAMENTE los valores dentro de df["Card brand"].unique()

10. Si buscas "mayores a X pesos":
    SIEMPRE usa la columna "TPV".

11. NO inventes valores como "Retail", "linea1", "Mastercard" (incorrecto).
    Usa SOLO lo que existe en df.

12. NO generes c√≥digo de varias l√≠neas.
    TU RESPUESTA ES SOLO UNA L√çNEA.

13. NO agregues explicaciones, comentarios ni texto fuera del c√≥digo.

14. Para saber si una transacci√≥n es de cr√©dito o d√©bito,
    USA SIEMPRE la columna "Card type".
    NO uses la columna "Txn type".

15. No uses comillas ni backticks alrededor del c√≥digo.

Tu respuesta debe ser SOLO una expresi√≥n que pandas pueda evaluar.
"""

def ask_df(query, df):
    # Llamada a Ollama
    response = client.chat(
        model="qwen2.5:7b",
        messages=[
            {"role": "system", "content": SYSTEM_PROMPT},
            {"role": "user", "content": query},
        ]
    )

    # Extraer el c√≥digo
    code = response['message']['content'].strip()
    print("C√≥digo generado por Qwen:", code)

    # Ejecutar el c√≥digo de forma segura
    allowed_builtins = {
        "len": len,
        "sum": sum,
        "min": min,
        "max": max,
        "round": round,
        "abs": abs,
        "sorted": sorted,
    }
    
    safe_locals = {"df": df}
    
    try:
        result = eval(code, {"__builtins__": allowed_builtins}, safe_locals)
        return result
    except Exception as e:
        print("ERROR AL EJECUTAR:", code)
        return f"Error ejecutando el c√≥digo '{code}': {e}"
````

````python
import re

def clean_code(code):
    # Quitar triple comillas si las pone
    code = code.strip().strip("```python").strip("```")

    # Quitar comillas externas si existen
    if (code.startswith('"') and code.endswith('"')) or (code.startswith("'") and code.endswith("'")):
        code = code[1:-1]

    # Reemplazar \n por saltos reales
    code = code.replace("\\n", "\n")

    # Quitar comillas alrededor de todo el c√≥digo si las tiene
    code = re.sub(r'^[\'"](.*)[\'"]$', r'\1', code, flags=re.DOTALL)

    return code.strip()
````
